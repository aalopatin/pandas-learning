```python
import pandas as pd
import numpy as np
import timeit

def test_solutions(*solutions, number):
    for idx, solution in enumerate(solutions):
        def timed_solution():
            solution()
            
        time = timeit.timeit(timed_solution, number=number)
        print(f"Solution {idx + 1} Time: {time:.6f} seconds")

def test_solutions_1(*solutions, df, number):
    for idx, solution in enumerate(solutions):
        def timed_solution():
            solution(df.copy())
            
        time = timeit.timeit(timed_solution, number=number)
        print(f"Solution {idx + 1} Time: {time:.6f} seconds")

def test_solutions_3(*solutions, df1, df2, number):
    for idx, solution in enumerate(solutions):
        def timed_solution():
            solution(df1.copy(), df2.copy())
            
        time = timeit.timeit(timed_solution, number=number)
        print(f"Solution {idx + 1} Time: {time:.6f} seconds")
```

```python
# Sample DataFrame
data = {'Category': ['A', 'A', 'B', 'B', 'C', 'C'],
        'Value': [1, 2, 3, 4, 5, 6]}
df = pd.DataFrame(data)
print("Sample DataFrame:")
print(df)
# Group by 'Category'
grouped = df.groupby('Category')
# Filter groups where the sum of 'Value' > 5
print("\nFilter groups where the sum of 'Value' > 5")
filtered = grouped.filter(lambda x: x['Value'].sum() > 5)

print(filtered)
```

##### 14. GroupBy and Handling Missing data:
Write a Pandas program to handle missing data in GroupBy operations to ensure accurate and reliable data analysis.

```python
df = pd.read_csv('data/titanic.csv')
df = df.drop('Unnamed: 15', axis=1)
print('Original DataFrame:')
display(df)

print('Groupby WITH dropna')
display(df.dropna(subset=['age']).groupby(['class', 'sex']).agg({'age': 'mean'}))

print('Groupby WITHOUT dropna')
display(df.fillna(100).groupby(['class', 'sex']).agg({'age': 'mean'}))




```
