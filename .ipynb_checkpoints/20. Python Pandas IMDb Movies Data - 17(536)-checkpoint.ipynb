```python
import pandas as pd
import numpy as np
import timeit

def test_solutions(*solutions, number):
    for idx, solution in enumerate(solutions):
        def timed_solution():
            solution()
            
        time = timeit.timeit(timed_solution, number=number)
        print(f"Solution {idx + 1} Time: {time:.6f} seconds")

def test_solutions_1(*solutions, df, number):
    for idx, solution in enumerate(solutions):
        def timed_solution():
            solution(df.copy())
            
        time = timeit.timeit(timed_solution, number=number)
        print(f"Solution {idx + 1} Time: {time:.6f} seconds")

def test_solutions_3(*solutions, df1, df2, number):
    for idx, solution in enumerate(solutions):
        def timed_solution():
            solution(df1.copy(), df2.copy())
            
        time = timeit.timeit(timed_solution, number=number)
        print(f"Solution {idx + 1} Time: {time:.6f} seconds")
```

##### 14. Write a Pandas program to get the longest runtime and shortest runtime.

```python
df = pd.read_csv('data/movies_metadata.csv')
small_df = df[['title', 'release_date', 'budget', 'revenue', 'runtime']]
small_df[
    (small_df['runtime'] == small_df['runtime'].max()) |
    (small_df['runtime'] == small_df['runtime'].min())
]
```
