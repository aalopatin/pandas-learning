```python
import pandas as pd
import numpy as np
import timeit

def test_solutions(*solutions, df, number):
    for idx, solution in enumerate(solutions):
        def timed_solution():
            solution(df.copy())
            
        time = timeit.timeit(timed_solution, number=number)
        print(f"Solution {idx + 1} Time: {time:.6f} seconds")
```

```python
import pandas as pd
df = pd.DataFrame( {'id' : ['A','A','A','A','A','A','B','B','B','B','B'], 
                    'type' : [1,1,1,1,2,2,1,1,1,2,2], 
                    'book' : ['Math','Math','English','Physics','Math','English','Physics','English','Physics','English','English']})

print("Original DataFrame:")
print(df)
new_df = df[['id', 'type', 'book']].drop_duplicates()\
                         .groupby(['id','type'])['book']\
                         .apply(list)\
                         .reset_index()

new_df['book'] = new_df.apply(lambda x: (','.join([str(s) for s in x['book']])), axis = 1)
print("\nList all unique values in a group:")
print(new_df)
```

```python
df = pd.DataFrame({
    'id': ['A','A','A','A','A','A','B','B','B','B','B'],
    'type': [1,1,1,1,2,2,1,1,1,2,2],
    'book': ['Math', 'Math', 'English', 'Physics', 'Math', 'English', 'Physics', 'English', 'Physics', 'English', 'English']
})


def unique(x):
    return ','.join(set(x))

def solution_1(df):
    return df.groupby(['id', 'type']).agg({'book': set})

def solution_2(df):
    return df.groupby(['id', 'type'])['book'].apply(unique).reset_index()

def solution_3(df):
    return df.groupby(['id', 'type']).agg({'book': lambda x: ','.join(sorted(set(x)))}).reset_index()

def solution_4(df):
    return df.groupby(['id', 'type']).agg({'book': pd.unique}).reset_index()

def solution_5(df):
    new_df = df[['id', 'type', 'book']].drop_duplicates()\
                         .groupby(['id','type'])['book']\
                         .apply(list)\
                         .reset_index()

    new_df['book'] = new_df.apply(lambda x: (','.join([str(s) for s in x['book']])), axis = 1)
    return new_df

test_solutions(
    solution_1, 
    solution_2, 
    solution_3, 
    solution_4, 
    solution_5, 
    df=df, 
    number=10000
)
```
