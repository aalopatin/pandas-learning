```python
import pandas as pd
import numpy as np
import timeit

def test_solutions(*solutions, df, number):
    for idx, solution in enumerate(solutions):
        def timed_solution():
            solution(df.copy())
            
        time = timeit.timeit(timed_solution, number=number)
        print(f"Solution {idx + 1} Time: {time:.6f} seconds")
```

##### 6. Write a Pandas program to create a Pivot table and find survival rate by gender, age wise of various classes.

```python
pd.set_option("display.max_rows", 0)
df = pd.read_csv('data/titanic.csv')
result = df.pivot_table(index=['sex', 'age'], columns='class', values='survived').round(2)
display(result)
pd.reset_option("display.max_rows")
```

```python
df = pd.read_csv('data/titanic.csv')
labels = ['children', 'early working age', 'prime working age', 'mature working age', 'elderly']
df['age'] = pd.cut(df['age'], bins=[0, 14, 24, 54, 64, np.inf], labels=labels)

result = df.pivot_table(index=['sex', 'age'], columns='class', values='survived', aggfunc=['sum', 'count', 'mean'], observed=False)
result['mean'] = (result['mean'] * 100).round(2)

map_func = {
    'sum': 'Survivors',
    'count': 'Passengers',
    'mean': 'Survivors %'
}

result.columns = [f'{map_func[func]} ({cls} class)' for func, cls in result.columns]


display(result)
```

##### 10. Write a Pandas program to create a Pivot table and find survival rate by gender, age of the different categories of various classes. Add the fare as a dimension of columns and partition fare column into 2 categories based on the values present in fare columns.

```python
df = pd.read_csv('data/titanic.csv')
age = pd.cut(df['age'], bins=[0, 10, 30, 60, 80])
fare = pd.qcut(df['fare'], 2)
result = df.pivot_table(index=['sex', age], columns=['class', fare], values='survived', aggfunc='mean', observed=False)
display(result)
```
