```python
import pandas as pd
import numpy as np
import timeit

def test_solutions(*solutions, df, number):
    for idx, solution in enumerate(solutions):
        def timed_solution():
            solution(df.copy())
            
        time = timeit.timeit(timed_solution, number=number)
        print(f"Solution {idx + 1} Time: {time:.6f} seconds")
```

##### 1. Create a dataframe of ten rows, four columns with random values. Write a Pandas program to highlight the negative numbers red and positive numbers black.

```python
df = pd.concat([pd.DataFrame({'A': np.arange(1,11)}), pd.DataFrame(np.random.randn(10, 4), columns=list('BCDE'))], axis=1)
def color_positive_negative(val):
    if val > 0:
        color = 'black'
    else:
        color = 'red'
    return f'color: {color}'

df.style.map(color_positive_negative)
```

```python
np.random.seed(24)
df = pd.DataFrame({'A': np.linspace(1, 10, 10)})
df = pd.concat([df, pd.DataFrame(np.random.randn(10, 4), columns=list('BCDE'))],
               axis=1)

print("Original array:")
print(df)

def color_negative_red(val):
    color = 'red' if val < 0 else 'black'
    return 'color: %s' % color

print("\nNegative numbers red and positive numbers black:")
df.style.map(color_negative_red)
```

##### 2. Create a dataframe of ten rows, four columns with random values. Convert some values to nan values. Write a Pandas program which will highlight the nan values.

```python
data = np.random.randn(10, 4)
index_nan = np.random.choice(data.size, 5, replace=False)
data.ravel()[index_nan] = np.nan

df = pd.concat(
    [
        pd.DataFrame({'A': np.arange(1,11)}), 
        pd.DataFrame(data, columns=list('BCDE'))
    ], axis=1
)

df.style.highlight_null()
```

##### 3. Create a dataframe of ten rows, four columns with random values. Write a Pandas program to highlight the maximum value in each column.

```python
data = np.random.randn(10, 4)

df = pd.concat(
    [
        pd.DataFrame({'A': np.arange(1,11)}), 
        pd.DataFrame(data, columns=list('BCDE'))
    ], axis=1
)
df.style.highlight_max(color='green', subset=list('BCDE'))
```

```python
np.random.seed(24)
df = pd.DataFrame({'A': np.linspace(1, 10, 10)})
df = pd.concat([df, pd.DataFrame(np.random.randn(10, 4), columns=list('BCDE'))],
               axis=1)
df.iloc[0, 2] = np.nan
df.iloc[3, 3] = np.nan
df.iloc[4, 1] = np.nan
df.iloc[9, 4] = np.nan
print("Original array:")
print(df)
def highlight_max(s):
    '''
    highlight the maximum in a Series green.
    '''
    is_max = s == s.max()
    return ['background-color: green' if v else '' for v in is_max]

print("\nHighlight the maximum value in each column:")
df.style.apply(highlight_max,subset=pd.IndexSlice[:, ['B', 'C', 'D', 'E']])
```

```python
count = 100000
data = np.random.randn(100, count)

df = pd.DataFrame(data)

def solution_1(df):
    return df.style.highlight_max(color='green')

def solution_2(df):
    def highlight_max(s):
        is_max = s == s.max()
        return ['background-color: green' if v else '' for v in is_max]
    return df.style.apply(highlight_max)

test_solutions(solution_1, solution_2, df=df, number=1)
```

##### 10. Create a dataframe of ten rows, four columns with random values. Write a Pandas program to highlight the entire row in Yellow where a specific column value is greater than 0.5.

```python
data = np.random.randn(10, 4)

df = pd.concat(
    [
        pd.DataFrame({'A': np.arange(1,11)}), 
        pd.DataFrame(data, columns=list('BCDE'))
    ], axis=1
)

def highlight_greaterthan(x):
    if x.C > .5:
        return ['background-color: yellow']*5
    else:
        return ['background-color: white']*5 
df.style.apply(highlight_greaterthan, axis=1)
```
