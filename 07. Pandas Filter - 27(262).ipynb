```python
import pandas as pd
import numpy as np
import timeit

def test_solutions(*solutions, df):
    for idx, solution in enumerate(solutions):
        def timed_solution():
            solution(df.copy())
            
        time = timeit.timeit(timed_solution, number=1000)
        print(f"Solution {idx + 1} Time: {time:.6f} seconds")

df = pd.read_csv('data/world_alcohol.csv')
df
```

##### 21. Write a Pandas program to filter rows based on row numbers ended with 0, like 0, 10, 20, 30 from world alcohol consumption dataset.

```python
def solution_1(df):
    return df[df.index % 10 == 0]

def solution_2(df):
    return df.filter(regex='0$', axis=0)

test_solutions(solution_1, solution_2, df=df)
```
