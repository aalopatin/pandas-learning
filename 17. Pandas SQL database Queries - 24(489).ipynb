```python
import pandas as pd
import numpy as np
import timeit

def test_solutions(*solutions, number):
    for idx, solution in enumerate(solutions):
        def timed_solution():
            solution()
            
        time = timeit.timeit(timed_solution, number=number)
        print(f"Solution {idx + 1} Time: {time:.6f} seconds")

def test_solutions_1(*solutions, df, number):
    for idx, solution in enumerate(solutions):
        def timed_solution():
            solution(df.copy())
            
        time = timeit.timeit(timed_solution, number=number)
        print(f"Solution {idx + 1} Time: {time:.6f} seconds")

def test_solutions_3(*solutions, df1, df2, number):
    for idx, solution in enumerate(solutions):
        def timed_solution():
            solution(df1.copy(), df2.copy())
            
        time = timeit.timeit(timed_solution, number=number)
        print(f"Solution {idx + 1} Time: {time:.6f} seconds")
```

```python
import os
from pathlib import Path
from sqlalchemy import create_engine
import sqlite3

path = 'data/database'
dfs = {}
for file in Path(path).iterdir():
    name = file.stem.lower()
    dfs[name] = pd.read_csv(file.absolute())

engine = create_engine('sqlite:///:memory:')
for name, df in dfs.items():
    df.to_sql(name, con=engine, index=False, if_exists='replace')
```

##### 21. Write a Pandas program to display the ID for those employees who did two or more jobs in the past.

```python
query = """
SELECT
    employee_id
FROM
    job_history
GROUP BY
    employee_id
HAVING
    count(*) > 1
"""

result = pd.read_sql(query, con=engine)
display(result['employee_id'].values)
```

##### 24. Write a Pandas program to display the first and last name and date of joining of the employees who is either Sales Representative or Sales Man.

```python
query = """
SELECT
    first_name,
    last_name,
    hire_date
FROM
    employees e
    INNER JOIN jobs j ON 
    e.job_id = j.job_id
    AND j.job_title in ('Sales Representative', 'Sales Man')
"""

result = pd.read_sql(query, con=engine)
display(result.head())
```

```python
employees_query = """
SELECT
    *
FROM
    employees
"""

jobs_query = """
SELECT
    *
FROM
    jobs
"""

employees = pd.read_sql(employees_query, con=engine)
jobs = pd.read_sql(jobs_query, con=engine)

jobs_id =  jobs[jobs['job_title'].isin(('Sales Representative', 'Sales Man'))]['job_id'].values

result = employees[employees['job_id'].isin(jobs_id)]
display(result[['first_name', 'last_name', 'hire_date']].head())
```
