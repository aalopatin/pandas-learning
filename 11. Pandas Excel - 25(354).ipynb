```python
import pandas as pd
import numpy as np
import timeit

def test_solutions(*solutions, df, number):
    for idx, solution in enumerate(solutions):
        def timed_solution():
            solution(df.copy())
            
        time = timeit.timeit(timed_solution, number=number)
        print(f"Solution {idx + 1} Time: {time:.6f} seconds")
```

##### 22. Write a Pandas program to import three datasheets from a given  excel data (coalpublic2013.xlsx ) and combine in to a single dataframe.
Note: Structure of the three datasheets are same.

```python
df = pd.read_excel('data/employee.xlsx', sheet_name=[0,1,2])
pd.concat(df).reset_index(drop=True)
```

```python
df1 = pd.read_excel('data/employee.xlsx',sheet_name=0)
df2 = pd.read_excel('data/employee.xlsx',sheet_name=1)
df3 = pd.read_excel('data/employee.xlsx',sheet_name=2)
df = pd.concat([df1, df2, df3]).reset_index(drop=True)
print(df)
```

```python
df = pd.concat([pd.read_excel('data/employee.xlsx')])

def solution_1(df):
    df = pd.read_excel('data/employee.xlsx', sheet_name=[0,1,2])
    return pd.concat(df).reset_index(drop=True)

def solution_2(df):
    df1 = pd.read_excel('data/employee.xlsx',sheet_name=0)
    df2 = pd.read_excel('data/employee.xlsx',sheet_name=1)
    df3 = pd.read_excel('data/employee.xlsx',sheet_name=2)
    return pd.concat([df1, df2, df3]).reset_index(drop=True)

test_solutions(solution_1, solution_2, df=df, number=1)
```

##### 25. Write a Pandas program to import three datasheets from a given excel data (employee.xlsx ) into a single dataframe and export the result into new Excel file.
Note: Structure of the three datasheets are same.

```python
df = pd.concat(pd.read_excel('data/employee.xlsx', sheet_name=[0,1,2])).reset_index(drop=True)

df['hire_date'] = pd.to_datetime(df['hire_date']).dt.date

with pd.ExcelWriter('data/employee_new.xlsx', date_format='YYYY-MM-DD') as writer:
    df.to_excel(writer, index=False)
```
