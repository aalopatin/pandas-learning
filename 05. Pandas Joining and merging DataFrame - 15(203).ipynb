##### 15. Write a Pandas program to Combine two DataFrame objects by filling null values in one DataFrame with non-null values from other DataFrame.

```python
df1 = pd.DataFrame({
    'A': [np.nan, 0., np.nan],
    'B': [3, 4, 5]
})

df2 = pd.DataFrame({
    'A': [1, 1, 3],
    'B': [3., np.nan, 3.]
})

df1[df1.isnull()] = df2
display(df1)

df2[df2.isnull()] = df1
display(df2)
```

```python
df1 = pd.DataFrame({
    'A': [np.nan, 0., np.nan, 1],
    'B': [3, 4, 5, 1]
})

df2 = pd.DataFrame({
    'A': [1, 1, 3, 1],
    'B': [3., np.nan, 3., 1]
})

df1 = df1.combine_first(df2)
display(df1)

df2 = df2.combine_first(df1)
display(df2)
```

```python
df1_1 = pd.DataFrame({
    'A': [np.nan, 0., np.nan, 1],
    'B': [3, 4, 5, 6]
})

df2_1 = pd.DataFrame({
    'A': [1, 1, 3],
    'B': [3., np.nan, 3.]
})

df1_2 = pd.DataFrame({
    'A': [np.nan, 0., np.nan, 1],
    'B': [3, 4, 5, 6]
})

df2_2 = pd.DataFrame({
    'A': [1, 1, 3],
    'B': [3., np.nan, 3.]
})

def solution_1(df):
    df1_1[df1_1.isnull()] = df2_1
    return df1_1

def solution_2(df):    
    return df1_2.combine_first(df2_2)

df = pd.DataFrame()
display(solution_1(df))
display(solution_2(df))

test_solutions(solution_1, solution_2, df=df)
```
